```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
#direct Graph
G = nx.DiGraph()

#add nodes
G.add_nodes_from(["A","B","C","D"])

#add edges
G.add_edges_from([("A","B"),("B","C"),("C","A"),("A","D")])

#draw Graph
nx.draw(G,with_labels=True,node_color='lightblue')
```

```python
#in degree
print("In Degree of each node:")
for node, deg in G.in_degree():
  print(f"Node {node}: {deg}")


#out Dgree
print("Out Degree of each node:")
for node, deg in G.out_degree():
  print(f"Node {node}: {deg}")

#total dgree
print("Total Degree of each node:")
for node, deg in G.degree():
  print(f"Node {node}: {deg}")
```

```python
node = "A"
predecessors = list(G.predecessors(node))
print(f"Predecessors of node {node}: {predecessors}")
```

```python
#Successor of nodes
print("Successor of each node:")
for node in G.nodes():
  successors = list(G.successors(node))
  print(f"Node {node}: {successors}")

```

```python
adj_matrix = nx.adjacency_matrix(G).todense()
print("Adjacency Matrix:")
print(adj_matrix)
```

```python
import pandas as pd
```

```python
adj_df = pd.DataFrame(adj_matrix,columns=G.nodes(),index=G.nodes())
print("Adjacency Matrix as DataFrame:")
print(adj_df)
```

```python
#Shortest path between edges
#length between edges

source = "A"
target = "D"
shortest_path = nx.shortest_path(G,source=source,target=target)
shortest_path_length = nx.shortest_path_length(G,source=source,target=target)
print(f"Shortest path length between {source} and {target}: {shortest_path_length}")
print(f"Shortest path between {source} and {target}: {shortest_path}")
```
